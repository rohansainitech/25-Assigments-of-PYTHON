```python
# Q-4. Write a Python program to swap two variables?

a = int(input("enter the first variable: "))
b = int(input("enter the value of second variable: "))
print("Before swapping: a =", a, ", b =", b)

a, b = b, a

print("After swapping: a =", a, ", b =", b)
```
