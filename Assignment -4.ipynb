```python
# Q- 1. Write a Python Program to Find the Factorial of a Number?

num = int(input("Enter a number: "))

factorial = 1

if num < 0:
    print("Factorial does not exist for negative numbers.")
elif num == 0:
    print("The factorial of 0 is 1.")
else:
    for i in range(1, num + 1):
        factorial = factorial * i
    print("The factorial of", num, "is", factorial)
```

```python
# Q- 4 Write a Python Program to Check Armstrong Number?
# Armstrong number is a number whose sum of digits each raised to the power of number of digits equals the original number.

num = int(input("Enter a number: "))

# Preserve the original number
original = num

# Count digits
digits = len(str(num))

sum_of_powers = 0

while num > 0:
    digit = num % 10
    sum_of_powers += digit ** digits
    num //= 10

if sum_of_powers == original:
    print(original, "is an Armstrong number.")
else:
    print(original, "is not an Armstrong number.")
```

```python
# Q-5. Write a Python Program to Find Armstrong Number in an Interval?

lower = int(input("Enter lower range: "))
upper = int(input("Enter upper range: "))

print("Armstrong numbers between", lower, "and", upper, "are:")

for num in range(lower, upper + 1):
    digits = len(str(num))
    sum_of_powers = 0
    temp = num

    while temp > 0:
        digit = temp % 10
        sum_of_powers += digit ** digits
        temp //= 10

    if sum_of_powers == num:
        print(num)
```
